```python
import pandas as pd
import numpy as np
import datetime
from pandasql import sqldf
pysql = lambda q: sqldf(q, globals())
```

```python
df = pd.read_csv('NYS_traffic.csv')
df.head()
```

```python
df['new'] = df.apply(lambda row:datetime.datetime(row[0]//100,row[0]%100,1),axis=1)
df.head()
df.columns = [c.replace(' ', '_') for c in df.columns]
dx = pysql("SELECT new as Date,GEO_Summary as Type,sum(Passenger_Count) as Count FROM df group by new,GEO_Summary")
print(dx.head())
dp = pysql("SELECT Date,Count from dx where dx.Type='Domestic'")
dq = pysql("SELECT Date,Count from dx where dx.Type='International'")
print(dp.head())
print(dq.head())
dy = pysql("SELECT new as Date,sum(Passenger_Count) as Count FROM df group by new")
dy.to_csv('NYS_mod_tc.csv')
```
