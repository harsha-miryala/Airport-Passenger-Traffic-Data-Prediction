```python
import pandas as pd
import numpy as np
import datetime
```

```python
ts = pd.read_csv('forecast_timeseries (2).csv')
ts.head(12)
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
ts['Ensemble'] = (ts['ARIMA']+ts['HOLTZ']+ts['Econometric'])/3.0
ts['Ensemble2'] = (ts['ARIMA']+ts['Econometric'])/2.0
```

```python
rms = sqrt(mean_squared_error(ts['Actual'], ts['ARIMA']))
print('RMS error of ARIMA is : ',rms)
print('Normalised RMS error of ARIMA is : ',rms/ts['Actual'].mean())
rms = sqrt(mean_squared_error(ts['Actual'], ts['HOLTZ']))
print('RMS error of HOLTZ is : ',rms)
print('Normalised RMS error of HOLTZ is : ',rms/ts['Actual'].mean())
rms = sqrt(mean_squared_error(ts['Actual'], ts['Econometric']))
print('RMS error of Econometric is : ',rms)
print('Normalised RMS error of Econometric is : ',rms/ts['Actual'].mean())
rms = sqrt(mean_squared_error(ts['Actual'], ts['Ensemble']))
print('RMS error of Ensemble is : ',rms)
print('Normalised RMS error of Ensemble is : ',rms/ts['Actual'].mean())
rms = sqrt(mean_squared_error(ts['Actual'], ts['Ensemble2']))
print('RMS error of Ensemble2 is : ',rms)
print('Normalised RMS error of Ensemble2 is : ',rms/ts['Actual'].mean())
```
